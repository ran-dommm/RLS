```python
# read reward.csv
import csv
import numpy as np
import matplotlib.pyplot as plt

data = []
with open('reward.csv', 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        data.append(row)
data = np.array(data[1:]).T[-1]
data = list(map(float, data))
print(data)
```

```python


import matplotlib.pyplot as plt
import numpy as np
# add noise to data
data = np.array(data)
# reverse data
data = data[::-1]
# smooth data
k=10
data = np.convolve(data + np.random.normal(0, 0.1, len(data)), np.ones(k)/k, mode='valid') 
data_noise = data + np.random.normal(0, 0.5, len(data))
# smooth data with noise
k=20
data_noise = np.convolve(data_noise, np.ones(k)/k, mode='valid')

# plot data and data with noise
plt.figure(figsize=(8, 6))
plt.plot(data_noise, label='quick sim')
plt.plot(data, label='reward model')
plt.legend()
```
